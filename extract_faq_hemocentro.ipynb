```python
from bs4 import BeautifulSoup

with open("data/hemocentro.html", "r", encoding="utf-8") as fp:
    hmtl_text = fp.read()
soup = BeautifulSoup(hmtl_text)
```

```python
items = soup.find_all("div", {"class":"elementor-accordion-item"})
items
```

```python

import re

def remove_enum(text):
    return re.sub("[0-9]{1,2}\s–\s", "", text)
```

```python
faq = [
    {
        "question": remove_enum(item.find("a", {"class":"elementor-accordion-title"}).text), 
        "answer": item.find("div", {"class":"elementor-tab-content"}).text
        } for item in items
    ]
```

```python
faq
```

```python
import json
with open('data/hemocentro.json', 'w', encoding='utf-8') as fp:
    json.dump({"data":faq}, fp, indent=4)
```
